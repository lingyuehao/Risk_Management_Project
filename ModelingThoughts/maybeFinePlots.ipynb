```python
import pandas as pd
import matplotlib.pyplot as plt
df_model = pd.read_csv("data/processed/accepted_model_base.csv")
print(df_model.shape)
```

```python
sample_df = df_model.sample(min(5000, len(df_model)), random_state=42)

plt.figure(figsize=(8, 6))
for label in [0, 1]:
    subset = sample_df[sample_df["target_bad"] == label]
    plt.scatter(subset["fico_avg"], subset["int_rate"], alpha=0.4, s=10, label=f"target_bad={label}")

plt.xlabel("Average FICO")
plt.ylabel("Interest Rate")
plt.title("FICO vs Interest Rate by Outcome")
plt.legend()
plt.show()
```
