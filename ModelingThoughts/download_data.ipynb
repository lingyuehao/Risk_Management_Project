```python
from pathlib import Path
import shutil
import kagglehub

def find_file(root: Path, keywords):
    for f in root.rglob("*"):
        if f.is_file():
            name = f.name.lower()
            if all(k.lower() in name for k in keywords):
                return f
    return None

dataset_path = Path(kagglehub.dataset_download("wordsforthewise/lending-club"))
print("Kaggle cache path:", dataset_path)

accepted_file = find_file(dataset_path, ["accepted"])
rejected_file = find_file(dataset_path, ["rejected"])

print("Found accepted in cache:", accepted_file)
print("Found rejected in cache:", rejected_file)

raw_dir = Path(".").resolve() / "data" / "raw" / "lendingclub"
raw_dir.mkdir(parents=True, exist_ok=True)

if accepted_file is not None:
    accepted_dest = raw_dir / accepted_file.name
    shutil.copy2(accepted_file, accepted_dest)
    print("Copied accepted to:", accepted_dest)

if rejected_file is not None:
    rejected_dest = raw_dir / rejected_file.name
    shutil.copy2(rejected_file, rejected_dest)
    print("Copied rejected to:", rejected_dest)

print("\nNow in raw_dir:")
for f in raw_dir.iterdir():
    print("-", f.name)
```
